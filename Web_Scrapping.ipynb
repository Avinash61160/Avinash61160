```python
from bs4 import BeautifulSoup 

import requests 

import csv

import pandas as pd
```

```python
url="https://www.flipkart.com/apple-iphone-13-pro-max-graphite-1-tb/p/itmdbb23fa63b708?pid=MOBG6VF56FGKA8SG&lid=LSTMOBG6VF56FGKA8SGGXC16M&marketplace=FLIPKART&q=13+pro+max&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_ps&fm=search-autosuggest&iid=898de514-2187-4eb4-ba89-870b375410d2.MOBG6VF56FGKA8SG.SEARCH&ppt=sp&ppn=sp&ssid=lhg08qo4ao0000001663608722724&qH=65a006a96a091b00"

response = requests.get(url)
```

```python
response
```

```python
htmlcontent = response.content
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")

print(soup.prettify)
```

```python


review=soup.find('div',attrs={'class':'_3HKIdy'})

print(review.text)
```

```python
ratings = []

for a in soup.findAll('div', attrs={'class':'_3LWZlK _1BLPMq'}):
  rate = a

  ratings.append(rate.text)
```

```python
ratings
```

```python
rew_title = []

for a in soup.findAll('div', attrs={'class':'col _2wzgFH'}):
  name=a.find('p',attrs={'class':'_2-N8zT'})

  rew_title.append(name.text)
```

```python
rew_title
```

```python
full = []

for a in soup.findAll('div', attrs={'class':'t-ZTKy'}):
  full_review=a
  full.append(full_review.text)
```

```python
full
```

```python
df = pd.DataFrame({'Ratings':ratings,'Review Title':rew_title,'Full Review':full})

df.head()
```
